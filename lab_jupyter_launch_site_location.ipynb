Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
marker_cluster = MarkerCluster().add_to(site_map)
for index, row in spacex_df.iterrows():
    # Get launch record details
    site_name = row['Launch Site']
    lat, long = row['Lat'], row['Long']
    launch_outcome = row['class']
    marker_color = 'green' if launch_outcome == 1 else 'red'
    marker = folium.Marker(
        location=[lat, long],
        icon=folium.Icon(color=marker_color),
        popup=site_name
    )
    marker.add_to(marker_cluster)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians
data = {
    'Lat': [28.5623, 34.6328, 28.5733, 34.6156],
    'Long': [-80.5774, -120.6107, -80.6467, -120.5561],
    'class': [1, 0, 1, 0]
}
spacex_df = pd.DataFrame(data)
def assign_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=6)
marker_cluster = MarkerCluster().add_to(site_map)
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color'])
    )
    marker_cluster.add_child(marker)
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
coastline_lat = 28.56367
coastline_lon = -80.57163
launch_site_lat = spacex_df.loc[0, 'Lat']
launch_site_lon = spacex_df.loc[0, 'Long']
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
coastline_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
    )
)
marker_cluster.add_child(coastline_marker)
lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], weight=1)
site_map.add_child(lines)
site_map
```

```python
from math import radians, sin, cos, sqrt, atan2

# Sample DataFrame (replace with your actual DataFrame)
data = {
    'Launch Site': ['Launch Site A', 'Launch Site B', 'Launch Site C'],
    'Lat': [28.5623, 34.6328, 28.5733],
    'Long': [-80.5774, -120.6107, -80.6467],
    'Railway Lat': [28.56367, 34.6356, 28.5743],   # Replace with actual railway coordinates
    'Railway Long': [-80.57163, -120.6123, -80.6489],  # Replace with actual railway coordinates
    'Highway Lat': [28.56, 34.63, 28.57],   # Replace with actual highway coordinates
    'Highway Long': [-80.58, -120.61, -80.64],  # Replace with actual highway coordinates
    'Coast Lat': [28.55, 34.63, 28.57],   # Replace with actual coastline coordinates
    'Coast Long': [-80.59, -120.60, -80.65]  # Replace with actual coastline coordinates
}
launch_sites_df = pd.DataFrame(data)

# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # approximate radius of Earth in km
    
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    return distance

# Create a Folium map centered around an average location of launch sites
site_map = folium.Map(location=[launch_sites_df['Lat'].mean(), launch_sites_df['Long'].mean()], zoom_start=6)

# Iterate through launch sites DataFrame
for index, row in launch_sites_df.iterrows():
    # Calculate distances to nearest railway, highway, and coastline
    distance_to_railway = calculate_distance(row['Lat'], row['Long'], row['Railway Lat'], row['Railway Long'])
    distance_to_highway = calculate_distance(row['Lat'], row['Long'], row['Highway Lat'], row['Highway Long'])
    distance_to_coast = calculate_distance(row['Lat'], row['Long'], row['Coast Lat'], row['Coast Long'])
    
    # Create markers for railway, highway, and coastline
    railway_marker = folium.Marker(
        location=[row['Railway Lat'], row['Railway Long']],
        icon=folium.Icon(color='blue'),
        popup=f"Distance to railway: {distance_to_railway:.2f} km"
    )
    site_map.add_child(railway_marker)
    
    highway_marker = folium.Marker(
        location=[row['Highway Lat'], row['Highway Long']],
        icon=folium.Icon(color='orange'),
        popup=f"Distance to highway: {distance_to_highway:.2f} km"
    )
    site_map.add_child(highway_marker)
    
    coast_marker = folium.Marker(
        location=[row['Coast Lat'], row['Coast Long']],
        icon=folium.Icon(color='green'),
        popup=f"Distance to coastline: {distance_to_coast:.2f} km"
    )
    site_map.add_child(coast_marker)
    
    # Draw polylines from launch site to railway, highway, and coastline
    railway_polyline = folium.PolyLine(locations=[[row['Lat'], row['Long']], [row['Railway Lat'], row['Railway Long']]], color='blue')
    site_map.add_child(railway_polyline)
    
    highway_polyline = folium.PolyLine(locations=[[row['Lat'], row['Long']], [row['Highway Lat'], row['Highway Long']]], color='orange')
    site_map.add_child(highway_polyline)
    
    coast_polyline = folium.PolyLine(locations=[[row['Lat'], row['Long']], [row['Coast Lat'], row['Coast Long']]], color='green')
    site_map.add_child(coast_polyline)

# Display the map
site_map
```
